## About

1. Prioritas Scrapping dari https://www.dpr.go.id/uu/prolegnas merupakan RUU prioritas

undefined. Prolegnas Per Link dari link khusus setiap RUU contoh https://www.dpr.go.id/uu/detail/id/239

undefined. Prolegnas Main Page dari https://www.dpr.go.id/uu/prolegnas-long-list merupakan gambaran keseluruhan RUU yang dicanangkan pada suatu periode

Ketiga data memiliki perbedaan informasi yang disajikanserta keterbaruhuan yang berbeda, misal prolegnas per link ada data pengusul sedangkan perolegnas main page ada informasi tanggal terbaru update dari RUU. Selain itu informasi progress dari setiap sumber juga berbeda, sumber main page dan per link cenderung yang paling terbaru. Dalam data Prolegnas Combine, progress yang paling tinggi yang akan diambil dan informasi dari main page serta per link digabung.

Tambahan Relevan Informasi:

- Pengkategorian RUU berdasarkan Topik; kareana di dpr diajukan judul dahulu

- Ada RUU yang muncul di dua periode

- Ada RUU yg jadi pioritas di kedua periode, ada jg yang hanya merupakan prioritas di salah satunya, contoh prioritas tahun 2019 kemudianada di 2024 tapi bukan prioritas dan belum selesai -> RUU tersebut tetap dicatat sebagai prioritas

## Prioritas Scrapping

```
path_of_the_directory= '/work/Prioritas'
```

```
prioritas2024 = []
prioritas2019 = []
for filename in os.listdir(path_of_the_directory):
    fname = path_of_the_directory + '/' + filename
    f = open(fname, 'r')
    soup= BeautifulSoup(f.read(), 'html.parser')
    items = soup.find_all('tr', {'role':'row'})[1:-1]
    year = filename[:4]
    for item in items:
        name = item.find('a').text.strip()
        info = item.find_all('td')
        date = info[4].text.strip()
        sponsor = info[9].text.strip()
        status = info[5].text.strip()
        data = [name, date, sponsor, status, year]
        category = info[7].text
        if category == 'Prolegnas 2015-2019':
            prioritas2019.append(data)
        else:
            prioritas2024.append(data) 
```

```
df2019_prioritas_raw = pd.DataFrame(prioritas2019, columns=['Judul Ruu', 'Tanggal Diupdate', 'Pengusul', 'Status', 'Tahun Prioritas'])
df2019_prioritas_raw
```

## Prolegnas Scrapping per Link

```
def connectDPR(url):
    headers = {
        'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
    }
    req = requests.get(url, headers=headers)
    print(req)
    return BeautifulSoup(req.text, 'html.parser')
```

### Collecting all specific links of prolegnas 

```
url24 = 'https://www.dpr.go.id/uu/prolegnas-long-list'
soup = connectDPR(url24)
head = soup.find('tbody')
items = head.find_all('tr', 'role' == 'row')
```

```
links2024 = []
links2019 = []
main = 'https://www.dpr.go.id'
for item in items:
    category = item.find_all('td')[7].text.strip()
    address = item.find('a', href=True)
    if category == 'Prolegnas 2015-2019':
        links2019.append(main + address['href'])
    else:
        links2024.append(main + address['href'])
```

```
print(len(links2024))
print(len(links2019))
```

### Create a dataframe of prolegnas information from specific link

```
def recent_status(status_img, sponsors):
    if 'DPR' in sponsors:
        if ('grey' not in status_img[6]['src']):
            status = 'Selesai'
        elif ('grey' not in status_img[5]['src']):
            status = 'Keputusan'
        elif ('grey' not in status_img[4]['src']):
            status = 'Pembahasan'
        elif ('grey' not in status_img[3]['src']):
            status = 'Penetapan Usul'
        elif ('grey' not in status_img[2]['src']):
            status = 'Harmonisasi'
        elif ('grey' not in status_img[1]['src']):
            status = 'Penyusunan'
        else:
            status = 'Terdaftar'
    if 'PEMERINTAH' in sponsors:
        if ('grey' not in status_img[3]['src']):
            status = 'Selesai'
        elif ('grey' not in status_img[2]['src']):
            status = 'Keputusan'
        elif ('grey' not in status_img[1]['src']):
            status = 'Pembahasan'
        else:
            status = 'Terdaftar'
    if sponsors == 'DPD':
        if ('grey' not in status_img[4]['src']):
            status = 'Selesai'
        elif ('grey' not in status_img[3]['src']):
            status = 'Keputusan'
        elif ('grey' not in status_img[2]['src']):
            status = 'Pembahasan'
        elif ('grey' not in status_img[1]['src']):
            status = 'Penetapan Usul'
        else:
            status = 'Terdaftar'
    return status
```

```
def status_num(status):
    if status == 'Selesai':
        return 6
    elif status == 'Keputusan':
        return 5
    elif status == 'Pembahasan':
        return 4
    elif status == 'Penetapan Usul':
        return 3
    elif status == 'Harmonisasi':
        return 2
    elif status == 'Harmonisasi':
        return 1
    elif status == 'Terdaftar':
        return 0
    else:
        return -1

def higher_status(status1, status2):
    nstat1 = status_num(status1)
    nstat2 = status_num(status2)

    if nstat1>nstat2:
        return status1
    else:
        return status2
```

### 2024

```
list2024 = df2024_main['Judul Ruu'].tolist()
list2019 = df2019_main['Judul Ruu'].tolist()
```

```
data2024 = []
for i in list2024:
    name = i
    proposed = df2024_specific.loc[df2024_specific['Judul Ruu'] == name, 'Tanggal Diusul'].iloc[0]
    updated = df2024_main.loc[df2024_main['Judul Ruu'] == name, 'Tanggal Diupdate'].iloc[0]

    status1 = df2024_specific.loc[df2024_specific['Judul Ruu'] == name, 'Status'].item()
    status2 = df2024_main.loc[df2024_main['Judul Ruu'] == name, 'Status'].item()
    status = higher_status(status1, status2)

    sponsor = df2024_specific.loc[df2024_specific['Judul Ruu'] == name, 'Sponsor'].item()
    pengusul = df2024_specific.loc[df2024_specific['Judul Ruu'] == name, 'Pengusul'].item()
    penugasan = df2024_specific.loc[df2024_specific['Judul Ruu'] == name, 'Penugasan'].item()

    prioritas = df2024_specific.loc[df2024_specific['Judul Ruu'] == name, 'Prioritas'].item()
    prioritas_year = df2024_specific.loc[df2024_specific['Judul Ruu'] == name, 'Tahun Prioritas'].item()

    data = [name, proposed, updated, status, sponsor, pengusul, penugasan, prioritas, prioritas_year]
    data2024.append(data)

df2024_combine = pd.DataFrame(data2024, columns=['Judul Ruu', 'Tanggal Diusul', 'Tanggal Diupdate', 'Status', 'Sponsor', 'Pengusul', 'Penugasan', 'Prioritas', 'Tahun Prioritas'])
```
